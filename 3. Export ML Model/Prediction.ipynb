# Model for Predicting Weekly Walmart Sales

This interactive notebook lets users predict the weekly sales of a Walmart store given the parameters set by the users. It utilizes a pre-trained linear regression model fitted to a dataset of weekly Walmart sales from 2010 to 2012. It is important to note the model is not very accurate.

### How to use this notebook

First, run the first two cells to load the model and display the input UI. To make a prediction, adjust the parameters to your liking and run the third cell. It is not necessary to run the first two cells again in order to make a new prediction.

```python
import joblib
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

model = joblib.load("model/Walmart_Sales.joblib")
```

```python
layout = widgets.Layout(width='750px')
style = {'description_width': 'initial'}

temp_slider = widgets.FloatSlider(value=70, min=0, max=100, step=0.1,
                                  description="Temperature: (°F)",
                                  readout_format='.1f',
                                  layout=layout, style=style)
fuel_slider = widgets.FloatSlider(value=3, min=2.4, max=4.5, step=0.05,
                                  description="Fuel Price (USD):",
                                  layout=layout, style=style)
cpi_slider = widgets.FloatSlider(value=150, min=125, max=230, step=0.1,
                                 description="Consumer Price Index:",
                                 readout_format='.1f',
                                 layout=layout, style=style)
unemp_slider = widgets.FloatSlider(value=8, min=0, max=20, step=0.01,
                                  description="Unemployment Rate (%):",
                                  layout=layout, style=style)
month_dropdown = widgets.Dropdown(
    options=[('January', 1), ('February', 2), ('March', 3), ('April', 4),
             ('May', 5), ('June', 6), ('July', 7), ('August', 9),
             ('September', 9), ('October', 10), ('November', 11), ('December', 12)],
    value=1, description="Month:"
)
holiday_check = widgets.Checkbox(value=False, description="Set as Holiday Week")

display(temp_slider, fuel_slider, cpi_slider, unemp_slider, month_dropdown, holiday_check)
```

```python
temperature = temp_slider.value
fuel_price = fuel_slider.value
cpi = cpi_slider.value
unemployment = unemp_slider.value
month = month_dropdown.value
holiday_flag = holiday_check.value

print(f"Parameters:")
print(f"Temperature: {temperature}°F")
print(f"Fuel Price: ${fuel_price}")
print(f"CPI: {cpi}")
print(f"Unemployment Rate: {unemployment}%")
print(f"Month (1 to 12): {month}")
print(f"Holiday Flag set as {holiday_flag}\n")

features = {"Temperature": temperature, "Fuel_Price": fuel_price, "CPI": cpi,
            "Unemployment": unemployment, "Month": month, "Holiday_Flag": holiday_flag}
features_df = pd.DataFrame({"": features})

prediction = model.predict(features_df.T)
print(f"Predicted Weekly Sales: ${prediction[0]:.2f}")
```
